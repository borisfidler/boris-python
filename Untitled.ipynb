```python
#broj reci u recenici

train2['word_count'] = train2['skills'].apply(lambda x: len(str(x).split(" ")))
train2[['skills','word_count']].head()
```

```python
#broj karaktera

train2['char_count'] = train['skills'].str.len() ## this also includes spaces
train2[['skills','char_count']].head()
```

```python
#prosecna duzina reci

def avg_word(sentence):
  words = sentence.split()
  return (sum(len(word) for word in words)/len(words))

train2['avg_word'] = train2['skills'].apply(lambda x: avg_word(x))
train2[['skills','avg_word']].head()
```

```python
#to lowercase man

train2['skills'] = train2['skills'].apply(lambda x: " ".join(x.lower() for x in x.split()))
train2['skills'].head()
```

```python
train2['skills'] = train2['skills'].str.replace('[^\w\s]','')
train2['skills'].head()
```
